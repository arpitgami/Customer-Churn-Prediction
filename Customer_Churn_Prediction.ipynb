Data Preprocessing

```python
# label encoding of target column
df["Churn"]= df["Churn"].replace({"Yes":1,"No":0})
df["Churn"].value_counts()
```

```python
# label encoding of categorical features
# identifying columns with object data type
object_column = df.select_dtypes(include='object').columns
object_column
```

```python
# initialize a dictionary to save the encoders
encoders = {}
```

```python
# apply label encoding and store the encoders
for column in object_column:
  label_encoder = LabelEncoder()
  df[column]= label_encoder.fit_transform(df[column])
  encoders[column]= label_encoder
```

```python
# save the encoder to a pickle file
with open("encoders.pkl", "wb") as f:
  pickle.dump(encoders, f)

encoders
```

```python
df.head()
```

Training and test data split

```python
# splitting the features and target
X = df.drop(columns = ["Churn"])
Y = df["Churn"]
```

```python
# splitting the train and test data
X_train,X_test, Y_train, Y_test = train_test_split(X,Y,test_size=0.2,random_state=2)
print(Y_train.shape)
```

```python
print(Y_train.value_counts())
```

Synthetic Minority Oversampling Technique

```python
smote = SMOTE(random_state=2)
X_train_smote, Y_train_smote = smote.fit_resample(X_train,Y_train)
print(X_train_smote.shape)
```

```python
Y_train_smote.value_counts()
```

Model Training

```python
# training with default parameters
# dictonary of models
models = {
    "Decision Tree": DecisionTreeClassifier(random_state = 2),
    "Random Forest": RandomForestClassifier(random_state = 2),
    "XGBoost": XGBClassifier(random_state = 2)
}
```

```python
# dictionary to store the cross validation results
cv_scores = {}

# perform 5 fold cross validation for each model
for model_name, model in models.items():
  cv_score = cross_val_score(model, X_train_smote, Y_train_smote, cv = 5)
  cv_scores[model_name] = cv_score
  print(f"{model_name} CV Score: {np.mean(cv_score):.2f}")
```

```python
rfc = RandomForestClassifier(random_state = 2)
rfc.fit(X_train_smote, Y_train_smote)
```

Model Evaluation

```python
# calculate on test data
y_test_pred = rfc.predict(X_test)
print("Accuracy score:\n", accuracy_score(Y_test, y_test_pred))
print("Confusion Matrix:\n", confusion_matrix(Y_test, y_test_pred))
print("Classification Report:\n", classification_report(Y_test, y_test_pred))
```

```python
# save the trained model as a pickle file
model_data = {"model": rfc, "feature_names": X.columns.to_list()}
with open(" customer_churn_model.pkl", "wb") as f:
  pickle.dump(model_data, f)
```

Load the saved model and build a predictive system

```python
# load the saved model and feature name
with open(" customer_churn_model.pkl", "rb") as f:
  model_data = pickle.load(f)

loaded_model = model_data["model"]
feature_names = model_data["feature_names"]

print(loaded_model)
print(feature_names)
```

```python
input_data = {
    'gender': 'Female',
    'SeniorCitizen': 0,
    'Partner': 'Yes',
    'Dependents': 'No',
    'tenure': 1,
    'PhoneService': 'No',
    'MultipleLines': 'No phone service',
    'InternetService': 'DSL',
    'OnlineSecurity': 'No',
    'OnlineBackup': 'Yes',
    'DeviceProtection': 'No',
    'TechSupport': 'No',
    'StreamingTV': 'No',
    'StreamingMovies': 'No',
    'Contract': 'Month-to-month',
    'PaperlessBilling': 'Yes',
    'PaymentMethod': 'Electronic check',
    'MonthlyCharges': 29.85,
    'TotalCharges': 29.85,

}

input_data_df = pd.DataFrame([input_data])
input_data_df
```

```python
with open("encoders.pkl", "rb") as f:
  encoders = pickle.load(f)
```

```python
# encode categorical features using the saved encoders
for column, encoder in encoders.items():
  input_data_df[column] = encoder.transform(input_data_df[column])

input_data_df.head()
```

```python
# make a prediction
prediction = loaded_model.predict(input_data_df)
pred_prob = loaded_model.predict_proba(input_data_df)
print(prediction)

# results
print(f"Prediction: {'Churn' if prediction[0]==1 else 'No Churn'}")
print(f"Prediction Probability: {pred_prob}")
```
